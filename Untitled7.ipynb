```python
#importing pandas
import pandas as pd
import numpy as np
from pandas import Series
#creating dataframe
df=pd.DataFrame()
#reading dataframe
df=pd.read_csv("Heart_short.csv")
%timeit
df['AHD'] = df['AHD'].apply(lambda x: 0 if x=='No' else 1)
df
```

```python
#Normalizing the data

df['Age']=((df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min()))*1
df['RestBP']=((df['RestBP']-df['RestBP'].min())/(df['RestBP'].max()-df['RestBP'].min()))*1
df['Chol']=((df['Chol']-df['Chol'].min())/(df['Chol'].max()-df['Chol'].min()))*1
df['RestECG']=((df['RestECG']-df['RestECG'].min())/(df['RestECG'].max()-df['RestECG'].min()))*1
df['MaxHR']=((df['MaxHR']-df['MaxHR'].min())/(df['MaxHR'].max()-df['MaxHR'].min()))*1
df['Oldpeak']=((df['Oldpeak']-df['Oldpeak'].min())/(df['Oldpeak'].max()-df['Oldpeak'].min()))*1

df.head()
```

```python
feat_col=['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak']
X=df[feat_col]
print(X)
df['AHD']
#df.AHD[df.AHD=='Yes']=1
#df.AHD[df.AHD=='No']=0

y=df['AHD']
print(y)
```

```python
#Splitting the data

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split (X,y,test_size=0.25,random_state=3)
```

```python
#importing logistic Regression

from sklearn.linear_model import LogisticRegression

log_reg=LogisticRegression()
#training the model
log_reg.fit(X_train,y_train)
```

```python
#predicting on the testing data
pr_log_reg=log_reg.predict(X_test)
print(pr_log_reg)
```

```python
from sklearn.metrics import accuracy_score
#predicting the accuracy
acc_lr=accuracy_score(y_test,pr_log_reg)

print(acc_lr)

```

```python
lr_prob=log_reg.predict_proba(X_test)


from sklearn import metrics


fpr, tpr, thresholds = metrics.roc_curve(y_test, lr_prob[:,1], pos_label=1)

AUC = metrics.auc(fpr, tpr)
print(AUC)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.figure()

# ROC Curve
plt.plot(fpr, tpr, color='red', lw=2, 
         label='ROC Curve (area = %0.2f)' % AUC)

# Random Guess line
plt.plot([0, 1], [0, 1], color='blue', lw=1, linestyle='--')

# Range of X-Axis and Y-Axis
plt.xlim([-0.005, 1.005])
plt.ylim([0.0, 1.01])

# Titles , Labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")

plt.show()
```
