```python
#importing the library of pandas

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
from sklearn.model_selection import cross_val_score

df=pd.DataFrame()

df=pd.read_csv("Credit.csv")

df
```

```python
feat_col=['Income','Limit','Rating','Cards','Age','Education','Married']

X=df[feat_col]

print(X.shape)

X.head()
```

```python
y=df['Balance']

print(y[::20])
print(y.shape)
y.head()
```

```python
#Splitting the data into the test and training samples
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.24,random_state=4)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Using linear regression to train the model
lin_reg=LinearRegression()

lin_reg.fit(X_train,y_train)
```

```python
print(lin_reg.intercept_)

print(lin_reg.coef_)
```

```python
#predicting on the test samples
y_pred=lin_reg.predict(X_test)

print(y_pred)
```

```python
mse = metrics.mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

print(rmse)
```

```python
# using cross validation for the last part of B
mse_list=cross_val_score(lin_reg,X,y,cv=10,scoring='neg_mean_squared_error')

print(mse_list)
```

```python
mse_listp=-mse_list
rmse_list=np.sqrt(mse_listp)
print(rmse_list)
```

```python
print(rmse_list.mean())
```
