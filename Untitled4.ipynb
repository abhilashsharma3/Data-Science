```python
#importing pandas
import pandas as pd
```

```python
#Creating the dataframe
df=pd.DataFrame()
```

```python
#reading from file
df=pd.read_csv("Heart_s.csv")
```

```python
#printing the DataFrame
df
```

```python
feature_col=['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak']
X=df[feature_col]
print(X)
y=df['AHD']
y[::10]
```

```python
#importing the train_test_split from the sklearn to split the data

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

k = 3
knn_heart = KNeighborsClassifier(n_neighbors=k) 

#spliting the data into training and testing sample with random state=
X_train,X_test,y_train,y_test = train_test_split (X,y,test_size=0.25,random_state=5)

#training on training set
knn_heart.fit(X_train,y_train)


#importing the logistic regression and the Decision tree

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

#init the objects
log_reg=LogisticRegression()

dec_tree=DecisionTreeClassifier()

#Training the model
log_reg.fit(X_train,y_train)

dec_tree.fit(X_train,y_train)

#Predicting on a new data sample with input Age  RestBP  Chol  RestECG  MaxHR  Oldpeak

X_testing=[[45,125,187,2,178,1.6]]

y_predict_logreg=log_reg.predict(X_testing)

y_predict_dectree=dec_tree.predict(X_testing)

y_predict_knn=knn_heart.predict(X_testing)

#printing the prediction for the sample data
print(y_predict_logreg)

print(y_predict_dectree)

print(y_predict_knn)
```

```python
#predicting on the test sample using logistic regression
y_pre_lr=log_reg.predict(X_test)

#predicting on the test sample using decision tree
y_pre_dt=dec_tree.predict(X_test)

#prediciting on the test sample using knn
y_pre_knn=knn_heart.predict(X_test)
```

```python
#importing accuracy_score from sklearn to find accuracy of the trained model of both the logistic regression and the 
# decision tree

from sklearn.metrics import accuracy_score

acc_lr=accuracy_score(y_test,y_pre_lr)

acc_dt=accuracy_score(y_test,y_pre_dt)

acc_knn=accuracy_score(y_test,y_pre_knn)
```
