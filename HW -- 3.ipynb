# Question2: Debt Prediction
### in this, we work with 'Credit.csv' dataset from the textbook of "An Introduction to Statistical Learning."

```python
import pandas as pd  # import panda as pd
from sklearn.linear_model import LogisticRegression

my_logreg = LogisticRegression() # create an object of LogisticRegression "class".

df = pd.read_csv('C:/Users/afara/Documents/DS Lab/Credit.csv')
df # print dataset
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing 

y = df['Balance'] # select the target (last column) from the DataFrame
y.size
print(y)
```

```python
from sklearn.model_selection import train_test_split

feature_cols = ['Limit', 'Cards', 'Married']
X = df[feature_cols]

X = df[['Limit', 'Cards', 'Married']]

#target vector
y = df['Balance']

preprocessing.scale(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.24, random_state=4)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### Find out coefficients

```python
my_linreg = LinearRegression() #instantiated as an "object" of LinearRegression "class".

my_linreg.fit(X_train, y_train) # fitting the model to the training data:

print(my_linreg.intercept_) #print theta0
print(my_linreg.coef_)  # printing [Theta1, Theta2, Theta3]

y_prediction = my_linreg.predict(X_test)

print(y_prediction)
```

### RMSC (Root Mean Square Error)

```python
from sklearn import metrics
import numpy as np #import numpy

mse = metrics.mean_squared_error(y_test, y_prediction) # Calculating "Mean Square Error" (MSE):

rmse = np.sqrt(mse) #calculate "Root Mean Square Error" (RMSE)
print(rmse)
```

### 10-fold Cross-Validation

```python
from sklearn.model_selection import cross_val_score #importing the method:

feature_cols = ['Limit', 'Cards', 'Married']

X = df[feature_cols]

y = df['Balance'] 
print(X.shape)
print(y.shape)
```

```python
#"my_linreg" is 
my_linreg = LinearRegression() #instantiated an object of class(LinearRegression).

mse_list = cross_val_score(my_linreg, X, y, cv=10, scoring='neg_mean_squared_error')
print(mse_list)
```

```python
mse_list_positive = -mse_list  #"Mean Square Error" (MSE)
rmse_list = np.sqrt(mse_list_positive)
print(rmse_list)

print(rmse_list.mean())  # calculate the average RMSE as final result of cross validation.
```
